```python
# Dicionário onde cada produto será armazenado com seu respectivo preço
produtos = {}

print("=== Cadastro de Produtos ===")

while True:
    nome = input("Digite o nome do produto (ou pressione Enter para finalizar): ")

    # Caso o usuário apenas aperte Enter, o cadastro é encerrado
    if nome == "":
        break

    # Solicita o preço do produto e trata a vírgula como separador decimal
    preco_str = input("Digite o preço do produto: R$ ")
    preco = float(preco_str.replace(',', '.'))

    # Registra no dicionário usando o nome como chave
    produtos[nome] = preco

    print("Produto registrado com sucesso!\n")

# Exibição dos itens cadastrados
print("\nProdutos cadastrados:")
for item, valor in produtos.items():
    print(f"- {item}: R$ {valor:.2f}")
```
