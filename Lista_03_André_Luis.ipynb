```python
# 11 Elabore um algoritmo que leia um número de entrada que indicará a quantidade de números a serem lidos. Em seguida, leia números( conforme o valor mostrado anteriormente) e imprima a soma e a média aritmética dos números informados.

# Solicita ao usuário a quantidade de números que serão informados
quantidade = int(input("Digite a quantidade de números a serem informados: "))

# Inicializa a soma
soma = 0

# Loop para ler os números
for i in range(quantidade):
    numero = float(input(f"Digite o {i + 1}º número: "))
    soma += numero

# Calcula a média
media = soma / quantidade

# Imprime os resultados
print(f"Soma dos números: {soma}")
print(f"Média aritmética: {media}")
```

```python
"""12Elabore um algoritmo que leia um número de entrada que indicará a quantidade deregistros a serem lidos (N). Em seguida algoritmo deve solicitar o nome e idade de Npessoas e ao final apresentar o nome da pessoa mais velha.
"""
# Solicita a quantidade de registros a serem lidos
N = int(input("Digite a quantidade de pessoas: "))

# Inicializa variáveis para armazenar a pessoa mais velha
nome_mais_velho = ""
idade_mais_velha = -1  # Começa com valor impossível

# Loop para ler os dados das N pessoas
for i in range(N):
    nome = input(f"Digite o nome da pessoa {i+1}: ")
    idade = int(input(f"Digite a idade de {nome}: "))

    # Verifica se essa pessoa é a mais velha até agora
    if idade > idade_mais_velha:
        idade_mais_velha = idade
        nome_mais_velho = nome

# Mostra o resultado
print(f"A pessoa mais velha é {nome_mais_velho} com {idade_mais_velha} anos.")
```

```python
""13 Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.
"""
N = int(input("Informe a quantidade de registros: "))

somaM = somaF = 0
qtdM = qtdF = 0

for i in range(N):
    sexo = input(f"Informe o sexo da pessoa {i+1} (M/F): ").strip().upper()
    idade = int(input(f"Informe a idade da pessoa {i+1}: "))

    if sexo == "M":
        somaM += idade
        qtdM += 1
    elif sexo == "F":
        somaF += idade
        qtdF += 1

if qtdM > 0:
    print("Média de idade dos homens:", somaM / qtdM)
else:
    print("Não houve registro de homens.")

if qtdF > 0:
    print("Média de idade das mulheres:", somaF / qtdF)
else:
    print("Não houve registro de mulheres.")
```

```python
"""14 Elabore um algoritmo que solicite ao usuário 10 números reais e ao final apresente o
maior e o menor deles.
"""

maior = float('-inf')
menor = float('inf')

for i in range(10):
    num = float(input(f"Digite o {i+1}º número: "))

    if num > maior:
        maior = num
    if num < menor:
        menor = num

print("Maior número:", maior)
print("Menor número:", menor)
```

```python
"""15Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
nulo 0 (zero) o programa ordene e mostre todos os números informados de forma cresce.
"""

numeros = []

while True:
    num = float(input("Digite um número real (0 para parar): "))
    if num == 0:
        break
    numeros.append(num)

numeros.sort()  # ordena em ordem crescente

print("Números em ordem crescente:", numeros)
```

```python

'''
16Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.
'''
idades = []

while True:
    idade = int(input("Digite a idade do aluno: "))
    idades.append(idade)
    resposta = input("Deseja informar a idade do próximo aluno? (S/N) ").upper()
    if resposta != "S":
        break

mais_novo = min(idades)
mais_velho = max(idades)
acima_18 = sum(1 for i in idades if i > 18)
ate_18 = sum(1 for i in idades if i <= 18)
media = sum(idades) / len(idades)

# Para calcular a mediana
idades.sort()
n = len(idades)
if n % 2 == 1:
    mediana = idades[n//2]
else:
    mediana = (idades[n//2 - 1] + idades[n//2]) / 2

print(f"Mais novo: {mais_novo}")
print(f"Mais velho: {mais_velho}")
print(f"Alunos acima de 18: {acima_18}")
print(f"Alunos com até 18: {ate_18}")
print(f"Média: {media}")
print(f"Mediana: {mediana}")
```
