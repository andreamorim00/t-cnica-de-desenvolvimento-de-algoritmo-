```python
# Contador de 1 a 100 com FOR e outro com WHILE, exibindo apenas números pares.


print("Contador de 1 a 100 exibindo só os números pares usando FOR:")

#Usando essa estrutura, eu indico como eu quero o range. Coloco de onde eu quero
#que comece(2, porque 1 é ímpar), até onde eu quero que vá(101, pois se colocar
#100, ele vai só até 0 99) e indico o passo(2) que garante que pule de 2 em dois,
#garantindo saída apenas de números pares.

for i in range(2, 101, 2):
  print(i)
  #daí é só printar i, já com os comandos estabelecidos.


print("Contador de 1 a 100 usando WHILE, exibindo somente números pares:")

#começor atribuindo 2 ao contador, indicando de onde começará a saída
contador = 2

#Enquanto o contador for menor ou igual a 100, o código continuará sendo
#imprimido:
while contador<=100:
  if contador % 2 == 0: #Aqui eu uso a função aritimética % para verificar se o
  #número dividido por 2 tem resto 0, ou seja, é um número par.
   print(contador) #vai verificar se é par e printar o contador pré estabelecido(2)
  contador += 1
  #Depois de ser executado a primeira "rodada" , uso +=1 para adicionar 1 ao
  #contador, até chegar em 100.
```
