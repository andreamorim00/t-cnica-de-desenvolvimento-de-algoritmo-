```python
alunos = []  # Lista onde serão guardados os nomes

print("=== Cadastro de Alunos ===")

while True:
    nome = input("Digite o nome do aluno(a) ou pressione Enter para encerrar: ")

    # Se o usuário apenas apertar Enter, o cadastro é finalizado
    if nome.strip() == "":
        break

    # Adiciona o nome informado na lista
    alunos.append(nome)

# Exibe todos os nomes cadastrados
print("\nAlunos cadastrados:")
for aluno in alunos:
    print("-", aluno)
```
