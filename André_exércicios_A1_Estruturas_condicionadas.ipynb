```python
# Programa para verificar se a pessoa pode entrar em um evento

print("=== Verificação de Idade para Evento ===")

idade = int(input("Digite sua idade: "))

if idade < 0:
    print("Idade inválida!")
elif idade < 16:
    print("Entrada não permitida para menores de 16 anos.")
elif idade < 18:
    print("Permitida a entrada somente acompanhado de um responsável.")
else:
    print("Entrada liberada! Boa diversão.")
```
